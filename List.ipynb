```python
# Given list of all runs made by Virat , print runs made in odd matches
```

```python
runs=[62,85,74,10,12,101,122,99,81,55]
runs
```

```python
for i in range(len(runs)):
    print(runs[i%2!=0])
''' # [i%2!=0]
- i % 2 != 0 checks if i is odd.
- In Python, True and False are actually integers (True == 1, False == 0).
- So runs[i%2!=0] is equivalent to:
- runs[0] when i is even (because i % 2 != 0 → False → 0)
- runs[1] when i is odd (because i % 2 != 0 → True → 1)
'''
```

```python
for i in range(len(runs)):
    if i%2!=0:
        print(runs[i])
```

```python
a=['a','b','c','d','e','c','a','c']
count=0
for i in range(len(a)):
    if a[i]=='c':
        count+=1
print(count)
```

```python
def count_ele(a,element):
    count1=0
    for i in range(len(a)):
        if a[i]==element:
            count1+=1
    print(element,count1)
```

```python
#Given an input integer N , Create a list of size n containing "even" or "odd"
n= int(input())
new_list=[]
for i in range(1,n+1):
    if i%2==0:
        new_list.append("even")
    else:new_list.append("odd")
print(new_list)
```
