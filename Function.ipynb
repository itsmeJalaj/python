# Scope

```python
# local scope -> variable defined inside function are accessible only inside the funciton
# Global scope -> variable outside th function are acessible everywhere
```

```python
a = 10
def func():
    print('local', a)
```

```python
func()
```

```python
print('global')
```

```python
func()
print('global')
```

```python
# if you try to modify a global variable inside the function , it creates a local variable and doesn't affect global variable
```

```python
x =55
def hi():
    return x
x+=10
print(hi())
```

```python
x=55
def hi():
    x= 100 # create a local varible using gloabal varaible : x+=10
    x+=10
    return x
print(hi())
```

```python
a =10
def func():
    global a
    a=11
    print('local',a)
func()
print('glabal',a)
```
