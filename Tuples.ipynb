```python
# iterating over a list of tuples
```

```python
list1=[(1,"A"),(2,"J"),(3,"AJ"),(4,"Ab"),(5,"M")]
for i in list1:
    print("Roll_No",i[0],"\t Name:",i[1])
```

```python
for i,j in list1:
    print("Roll_no :",i,"\t Name:",j)
```

```python
for roll_no, name in list1:
    print("Name",name,"roll no",roll_no)
```
