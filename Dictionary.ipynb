# comprehension

```python
## shortcut to create list/set/dictionary
# ex --> out=[i for i in range()]
```

```python
N= int(input())
new_list=[i for i in range(1,N+1)]
print(new_list)
```

```python
n=int(input())
new_list=[i**2 for i in range(1,n+1)]
print(new_list)
```

```python
# advantages of comprehension
# -> there is no such great advantage but 
#     -> saves us from writing few lines of code
#     -> sightly faster but not that much
```

```python
n= int(input())
new_list=[i for i in range(1, n+1) if i%2==0]
print(new_list)
```
