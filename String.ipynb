```python
# take a string input from user , return "Palindrome" if the stirng is palindrome . write case insenstive function
```

```python
str1= input()
for i in str1:
    print(i)
```

```python
str1= input()
for i in range(len(str1)-1,-1,-1):
    print(str1[i])
```

```python
str1= input().lower()
rev=''
for i in range(len(str1)-1,-1,-1):
    rev+= "".join(str1[i])
if rev==str1:
    print("palindrome")
else:
    print('Not palindrome')
```
